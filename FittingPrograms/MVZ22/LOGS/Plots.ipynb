```python
import os
import json
PATH="/data/arTeMiDe_Repository/DataProcessor/FittingPrograms/MVZ22/LOGS/"
fList=[f for f in os.listdir(PATH) if ".dat" in f]
fList.sort()
print(fList)
```

```python
import numpy as np
import matplotlib.pyplot as plt
fig, axs = plt.subplots(1, 3, figsize=(20, 5))
i=0

for fname in ['A8-00y04plt.dat','CDF1plt.dat','E228-400plt.dat']:
    with open(PATH+fname, 'r') as openfile:
        sDICT = json.load(openfile)
    
    kToPlot=[el for el in sDICT.keys() if not el in ["p","qT"]]
    
    edges=[0.]+[p[1] for p in sDICT["qT"]]
       
    
    for key in kToPlot:
        axs[i].stairs(sDICT[key], edges, label=key, baseline=None)        
        axs[i].legend()
        
    axs[i].title.set_text(fname[:-7])
    i+=1                         
    

plt.show()
```
